Se verifica los datos nulos de la base de datos, en este caso la columna que presenta datos faltantes es el person_emp_length y el loan_int_rate con 895 y 3116 datos nulos respectivamente

```python
datos_faltantes = df.isnull()
for columna in datos_faltantes.columns.values.tolist():
    print('columna:',columna)
    print (datos_faltantes[columna].value_counts())
    print("")    
```

Vamos a manipular los datos faltantes, se va a realizar las siguientes manipulaciones:
- La columna person_emp_length vamos a manipularlo, reemplazando los 895 datos faltantes por su promedio
- La columna loan_int_rate vamos a manipularlo, reemplazando los 3116 datos faltantes por su promedio

```python
person_emp_length_mean = df["person_emp_length"].astype("float").mean(axis=0)
person_emp_length_mean

loan_int_rate_mean = df["loan_int_rate"].astype("float").mean(axis=0)
loan_int_rate_mean

df["person_emp_length"].replace(np.nan, person_emp_length_mean, inplace=True)
df["loan_int_rate"].replace(np.nan, loan_int_rate_mean, inplace=True)
```

- Reemplazar los Valores (N,Y) en (0,1) respectivamente, estos estan presentes en la columna cb_person_default_on_file
- 

```python
df["cb_person_default_on_file"].replace('Y', 1, inplace=True)
df["cb_person_default_on_file"].replace('N', 0, inplace=True)

#Verificamos imprimiendo los primeros 10 valores si el reemplazo se ha realizado de manera correcta

df.head(10)
```
